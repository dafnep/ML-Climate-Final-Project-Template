```python
NAMES1 = ['ACCESS_CM2','CanESM5','CAMS_CSM1', 'CNRM_CM6_1','CNRM_ESM']
datasets1 = [ACCESS_CM2,CanESM5,CAMS_CSM1, CNRM_CM6_1,CNRM]
names2 = ['hadGEM3','MIROC6','UKESM1']
datasets2=[hadGEM31,MIRCO6_1,UKESM1_1]
datasets3=[hadGEM32,MIRCO6_2,UKESM1_2]
name=0
dictionnary = dict()
for d in datasets1:
    siconc=  d.variables['siconc'][:]
    smean = np.nanmean(siconc, axis =(1,2))
    sic = smean[11:-1]
    avgsic = np.mean(sic.reshape(-1, 3), axis=1)
    djf_t = []
    for i in range(0,656-4-3,4):
        djf_t.append(avgsic[i])
    dictionnary[NAMES1[name]] = djf_t
    name +=1
name = 0
for d1,d2 in zip(datasets2,datasets3):
    air1 = d1.variables['siconc'][:]
    air2 = d2.variables['siconc'][:]
    smean1 = np.nanmean(air1, axis =(1,2))
    smean2 = np.nanmean(air2, axis =(1,2))
    smean=np.hstack((smean1,smean2))
    sic = smean[11:-1]
    avgsic = np.mean(sic.reshape(-1, 3), axis=1)
    djf_t = []
    for i in range(0,656-4-3,4):
        djf_t.append(avgsic[i])
    dictionnary[names2[name]] = djf_t
    name+=1
```
